```python
# 0. Install required libraries
!pip uninstall duckduckgo-search -y
!pip install ddgs fastai fastdownload --quiet

# 1. Import libraries
from ddgs import DDGS
from fastai.vision.all import *
from pathlib import Path
import time

# 2. Define image search function with retry
def search_images(keywords, max_images=50, retries=3, delay=15):
    for attempt in range(retries):
        try:
            with DDGS() as ddgs:
                print(f"üîç Searching: {keywords} (Attempt {attempt+1})")
                results = ddgs.images(keywords, max_results=max_images)
                return L(results).itemgot('image')
        except Exception as e:
            print(f"‚ö†Ô∏è {e}")
            if attempt < retries - 1:
                print(f"‚è≥ Retrying in {delay} seconds...")
                time.sleep(delay)
    return []

# 3. Download and prepare dataset
categories = ['forest', 'bird']
path = Path('bird_or_not')

for cat in categories:
    dest = path/cat
    dest.mkdir(parents=True, exist_ok=True)
    urls = search_images(f'{cat} photo', max_images=50)
    print(f"üì• Found {len(urls)} images for '{cat}'")
    if urls:
        download_images(dest, urls=urls)
        time.sleep(10)
        resize_images(dest, max_size=400)

# 4. Remove corrupted images from all category folders
for cat in categories:
    folder = path/cat
    failed = verify_images(get_image_files(folder))
    print(f"üóëÔ∏è Removing {len(failed)} corrupted images from '{cat}'")
    failed.map(Path.unlink)

# 5. Create DataLoaders
dls = ImageDataLoaders.from_folder(
    path,
    train='.',
    valid_pct=0.2,
    seed=42,
    item_tfms=Resize(460),
    batch_tfms=aug_transforms(size=224)
)

# 6. Visualize sample batch
dls.show_batch(max_n=8)

# 7. Train model using transfer learning
learn = cnn_learner(dls, resnet18, metrics=accuracy)
learn.fine_tune(3)

# 8. Plot loss curve
learn.recorder.plot_loss()

# 9. Save model
learn.save('bird_not_bird_model')

# 10. Test prediction on a sample image
img_path = sorted((path/'bird').iterdir())[0]
pred, pred_idx, probs = learn.predict(img_path)
print(f"üïµÔ∏è Prediction: {pred}; Probability it's bird: {probs[pred_idx]:.4f}")
```
